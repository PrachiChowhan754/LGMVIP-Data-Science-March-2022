```python
#Importing required Libraries
import cv2
import matplotlib.pyplot as plt
```

```python
#Reading the Original Image
img = cv2.imread('img.jpg', 100)
plt.imshow(img)
```

```python
#Convert the RBG image to grayscale image
plt.imshow(img, cmap='gray')
```

```python
#Inverted the grayscale image
inv = 255-img
```

```python
plt.imshow(inv, cmap='gray')
```

```python
#Apply a Guassian Blur
blur = cv2.GaussianBlur(inv, ksize=(21,21), sigmaX=0, sigmaY=0)
```

```python
plt.imshow(blur, cmap='gray')
```

```python
#Mixing the gray scale image with inverted blurry image
dodge = lambda image, mask: cv2.divide(image, 255-mask, scale=256)
```

```python
blended = dodge(img, blur)
```

```python
plt.imshow(blended, cmap='gray')
```
